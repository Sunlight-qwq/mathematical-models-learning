## 灰色预测

尝试使用灰色预测的方法，对模型进行估计。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft
```

```python
df = pd.read_excel('data.xlsx', header=None)
data = df.values.flatten()
print(data.size)
plt.scatter(np.arange(104), data)
t = np.linspace(0, 103, 1000)
plt.plot(t, 4.8187 * np.exp(0.0316 * t))
```

```python
df_flat = pd.DataFrame(data)
df_flat.to_excel('data_flatten.xlsx')
```

指数回归

```python
r = 4.8187 * np.exp(0.0316 * np.arange(104)) - data
plt.scatter(np.arange(104), r)
```

```python
X = fft(r)
plt.scatter(np.arange(104), X)
```

将数据累加，计算级比。

```python
x1 = np.cumsum(data)
plt.scatter(np.arange(104), x1)
```

```python
n = 104
ratio_min = np.exp(-2 / (n + 1))
ratio_max = np.exp(2 / (n + 1))
ratios = [x1[k+1] / x1[k] for k in range(103)]
list(map(lambda t: ratio_min < t < ratio_max, ratios))
```

```python
x0 = data + 1000
ratios = [x0[k+1] / x0[k] for k in range(103)]
plt.scatter(np.arange(104), x0)
plt.ylim([0, 1200])
# plt.scatter(np.arange(103), list(map(lambda t: int(ratio_min < t < ratio_max), ratios)))
```
